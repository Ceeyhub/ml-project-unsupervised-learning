# Part II - KMeans Clustering

The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
from sklearn.cluster import KMeans
```

```python
#using the elbow method to look for various values of K to find where the rate of decrease sharply changes

inertias = []
K_range = range(1, 10)
for k in K_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df_selected_scaled)  # Using the preprocessed data
    inertias.append(kmeans.inertia_)
```

```python
#Ploting the result

plt.plot(K_range, inertias, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Choosing the optimal k based on the plot
chosen_k = 3
```

```python
# Perform K-means clustering
kmeans = KMeans(n_clusters=chosen_k)
kmeans.fit(df_selected_scaled)
labels = kmeans.predict(df_selected_scaled)
centroids = kmeans.cluster_centers_
```

```python
# Create a scatter plot to visualize clusters
plt.figure(figsize=(8, 6))
```

```python
# Plot data points colored by cluster labels
for cluster_label in range(chosen_k):
    plt.scatter(
        df_selected_scaled[labels == cluster_label, 0],  # X-axis (e.g., Fresh)
        df_selected_scaled[labels == cluster_label, 1],  # Y-axis (e.g., Milk)
        label=f'Cluster {cluster_label}'
    )
```

Even though 3 clusters can be seen, a distinct separation cannot be observed

```python
# Plot cluster centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker='x', s=100, label='Centroids')

plt.xlabel('Fresh')
plt.ylabel('Milk')
plt.title('K-means Clustering Results')
plt.legend()
plt.grid(True)
plt.show()
```

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

```python
# Perform PCA for dimensionality reduction
pca = PCA()
principal_components = pca.fit_transform(df_selected_scaled)
```

```python
# Variance explained by each principal component

explained_variance_ratio = pca.explained_variance_ratio_
```

```python
# Scree plot to visualize explained variance

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

```python
# Cumulative explained variance

cumulative_explained_variance = explained_variance_ratio.cumsum()
```

```python
# Determining the number of components to retain based on cumulative explained variance

desired_variance = 0.95  # threshold (95% of variance)
n_components_to_retain = (cumulative_explained_variance >= desired_variance).sum()
```

```python
# Retrain PCA with the selected number of components

pca = PCA(n_components=n_components_to_retain)
principal_components = pca.fit_transform(df_selected_scaled)
```

```python

# Create a scatter plot of the first two principal components
plt.scatter(principal_components[:, 0], principal_components[:, 1], c='b', marker='o', alpha=0.5)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Scatter Plot of First Two Components')
plt.grid(True)
plt.show()
```
